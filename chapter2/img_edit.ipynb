```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
a_img = cv.imread('a.png')
a_img = cv.cvtColor(a_img, cv.COLOR_BGR2RGB)
b_img = cv.imread('b.png')
b_img = cv.cvtColor(b_img, cv.COLOR_BGR2RGB)
c_img = cv.imread('c.png')
c_img = cv.cvtColor(c_img, cv.COLOR_BGR2RGB)

final_img = None
```
